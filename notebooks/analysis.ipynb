# AdventureLLM Analysis
Loads CSV logs from `data/raw_runs/` and computes simple aggregates.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

log_dir = Path('../data/raw_runs')
files = list(log_dir.glob('*.csv'))
files
```

```python
if not files:
    print('No log files found. Run an experiment first.')
else:
    df = pd.concat(pd.read_csv(f) for f in files)
    df['inventory_count'] = df['inventory'].fillna('').apply(lambda x: len([i for i in str(x).split(';') if i]))
    display(df.head())
    episode_summary = df.groupby(['run_id', 'episode_id', 'model_name']).agg({
        'score': 'last',
        'moves': 'max',
        'tokens_prompt': 'sum',
        'tokens_completion': 'sum',
    }).reset_index()
    display(episode_summary)
    plt.figure(figsize=(6,4))
    episode_summary.boxplot(column='score', by='model_name')
    plt.title('Final score distribution per model')
    plt.suptitle('')
    plt.xlabel('Model')
    plt.ylabel('Score')
    plt.show()
```
